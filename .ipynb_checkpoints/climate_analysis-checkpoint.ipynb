```python
# import dependencies

import sqlalchemy
from sqlalchemy import Column, Float, Integer, String
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, desc

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# connect to the sqlite database

engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# declare the base using automap

Base = automap_base()
```

```python
# reflect the database files

Base.prepare(engine, reflect=True)
```

```python
Base.classes.keys()
```

```python
# Choose a start date and end date for your trip (%Y-%m-%d)

vacay_start = '2018-07-01'
vacay_end = '2018-07-08'
```

```python
# create a session

session = Session(engine)
```

```python
# assign the classes to variables

measurement = Base.classes.Measurement
station = Base.classes.Station
```

# Precipitation Analysis

```python
# create a query to retrieve the last 12 months of precipitation data using the 'date' and 'prcp' columns

for row in session.query(measurement.date, measurement.prcp).filter(measurement.date > '2017-04-14').all():
    print(row)
```

```python
# load the query results into a Pandas DataFrame and set the index to the date column

precip = session.query(measurement.date, measurement.prcp).filter(measurement.date > '2017-04-14').\
        order_by(measurement.date).all()

precip_df = pd.DataFrame(precip)
precip_df = precip_df.set_index('date')
precip_df.head()
```

```python
# Plot the results using the DataFrame plot method

precip_df.plot.line(title="Precipitation Over Last 12 Months", color="blue", rot=45)
plt.show()
```

```python
# Use Pandas to print the summary statistics for the precipitation data
precip_df.describe()
```

# Station Analysis

```python
# query to calculate the total number of stations

station_num = session.query(station.station).count()
print(station_num)
```

```python
# query to find the most active stations
# List the stations and observation counts in descending order
# Which station has the highest number of observations?

station_count = session.query(func.count(measurement.station), measurement.station).group_by(measurement.station).\
    order_by(desc(func.count(measurement.station)))

for row in station_count:
    print(row)
```

station USC00519281, WAIHEE 837.5, HI US, has the most observations

```python
# query to retrieve the last 12 months of temperature observation data (tobs)

for row in session.query(measurement.date, measurement.tobs).filter(measurement.date > '2017-04-14').all():
    print(row)
```

```python
# Filter by the station with the highest number of observations.

waihee_temp = session.query(measurement.station, measurement.date, measurement.tobs).filter(measurement.station == 'USC00519281').\
            filter(measurement.date > '2017-04-14').all()
for row in waihee_temp:
    print(row)
```

```python
waihee_df = pd.DataFrame(waihee_temp)
waihee_df['tobs'].max()
waihee_df['tobs'].min()
```

```python
# Plot the results as a histogram with bins=12

bin = np.arange(60,84,2)
plt.hist(waihee_df['tobs'], bins=bin)
tick_locations = [value+0.9 for value in bin]
plt.xticks(tick_locations, bin)
plt.title('Temperature Frequency')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

# Temperature Analysis

```python
# Write a function called calc_temps that will accept a start date and end date (%Y-%m-%d) 
# and return the minimum, average, and maximum temperatures for that range of dates

def calc_temps():
    start_date = str(input("Enter your start date (%Y-%m-%d): "))
    end_date = str(input("Enter your end date (%Y-%m-%d): "))
    print(f"{start_date}, {end_date}")
    temp_min = session.query(func.min(measurement.tobs)).filter(measurement.date > start_date).\
            filter(measurement.date < end_date).all()
    temp_max = session.query(func.max(measurement.tobs)).filter(measurement.date > start_date).\
            filter(measurement.date < end_date).all()
    temp_avg = session.query(func.avg(measurement.tobs)).filter(measurement.date > start_date).\
            filter(measurement.date < end_date).all()
    print(temp_min)
    print(temp_max)
    print(temp_avg)

calc_temps()
```

```python
def daily_normals():
    date = str(input("Enter your chosen month and day (%m-%d): "))
    
    print(f"date chosen: {date}")

    sel = [ 
           func.min(measurement.tobs), 
           func.max(measurement.tobs), 
           func.avg(measurement.tobs)]
    norms = session.query(*sel).\
        filter(func.strftime("%m-%d", measurement.date) == date).all()

    return norms
daily_normals()
```
